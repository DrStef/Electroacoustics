<b> Simulation with WinISD </b>  <br>
    
<span style="color:#4169E1"> We validate our model with WinISD. <br>  WinISD by LinearTeam http://www.linearteam.org/ is an open source tool for simulating loudspeakers in various enclosures: closed, ported, passive radiator. Based on lumped parameters. <br>     
The Scilab/Matlab script was initially developped to design miniature speaker systems with passive radiators. WinISD is not well suited for miniature speakers and Thiele-Small parameters having very small values. Conversion are necessary and generally prone to errors.  <br>  

<span style="color:#4169E1"> WinISD model from [1].

<span style="color:#4169E1"> $P_{D+}$, $P_{D-}$ are the input "pressure" from the speaker. The speaker model is not represented here (see section  ). 

<span style="color:#4169E1"> $C_{CAB}$ and $R_{AB}$ are resp. the compliance and the damping of the enclosure. <br>
$R_{AL}$ is the resistor (damping) attached to an enclosure leak (generally a "slow leak" or "highly resistive leak").<br>
$C_{AR}$, $L_{MAR}$, $R_{AR}$ describe the suspension compliance, the mass and damping of the passive radiator. 



$$C_{AR} = C_{MS} S^2_D= \frac{V_{AS}}{ρc^2}$$

$$L_{MAR} = M_{MS} S^2_D$$ 

$$R_{AR} =\frac{1}{fsQ_{MS}C_{AR}}$$


$$R_{AL} = \frac{Q_L}{2πf_B C_{CAB}}$$

$$ R_{AB} =\frac{1}   {2πf_B Q_{AB}C_{CAB}}$$

<span style="color:#4169E1"> The model of a speaker enclosure with a passive radiator is a variant of the ported enclosure shown above. The "port" equivalent RL circuit must be replaced with a RLC circuit, with:
    
- <span style="color:#4169E1">Rpr: passive radiator damping (+acoustic radiated impedance: loss)
- <span style="color:#4169E1">Lpr: mass of the passive radiator cone (+ mass added by air)
- <span style="color:#4169E1">Cpr: compliance of passive radiator suspension.

#### Conclusions

<span style="color:#4169E1">  Overall, our model matches WinISD simulation pretty well except for damping under the passive radiator resonance. And at very low frequency the match is not perfect. <br>
Mismatch maybe due to:   
- <span style="color:#4169E1"> slow leak in the enclosure: "resistor" model (fluid viscosity).
- <span style="color:#4169E1"> damping in the enclosure: absorption model.
- <span style="color:#4169E1"> different model for acoustic impedance: added mass by the fluid (air), complex radiating impedance: resistor (loss). 
- <span style="color:#4169E1"> For computing diaphragm excursion xd we had to introduce a $\sqrt{2}$ term (peak displacement instead of average displacement ?). <br>
    
<span style="color:#4169E1"> Low frequency damping issues are not very consequential, because the present scripts are used to simulate miniature speaker systems, WinISD is generally not effective unless a unit conversion of all T/S parameters is implemented. The first resonance frequency $f_0$ of miniature system is relatively high, way above 100 Hz, and the impact of miniature enclosure damping is negligible.     
    
<span style="color:#4169E1"> In further works, we will analyse radiation impedance models and damping models, used by the team at LinearTeam compared with Leo Beranek's models.

```python
def parallel(z1,z2) :
    z = (z1*z2)/(z1+z2)
    return z

def series(z1,z2) :
    z = z1 + z2    
    return z
```

```python
#------------------
#Acoustic constants
#------------------
ro = 1.21  # Air density (kg/m3) 
cel = 344  # Sound celerity m/s
afreq= np.arange(1, 10000, 1, dtype=int) # frequency range for simluation
w = (2*np.pi)*afreq     # circular frequency (rad/s)
k = w/cel               # wavenumber rad/m              
```

```python
#============  Thiele Small Parameters (Electro-Mechanical Parameters) ==========================
#------- PEERLESS  830983  ---------
f0 = 142.27                        # 1st Resonance Frequency - free air (Hz)
w0 = (2*np.pi)*f0                  # circular frequency (rad/s)
Re = 4.093                         # Voice Coil DC resistance (Ohms)
Sd = 13e-04                        # Diaphragm Area (m2)
Vas = 200e-06                      # Equivalent volume (m3) - linked to suspension compliance
ksv = Sd**2*ro*cel**2/Vas   # Speaker suspension stiffness  (N/m)
#-----------------------------------------------

Mmd= ksv/((2*np.pi*f0)**2)

#Mmd = 1.5e-03        # Mass of speaker cone + voice coil  (kg)            
Bl = 2.8             # Force factor Bl(Tesla.m) 
# B= Magnetic flux density in air gap, l= length of voice coil wire 
# force on voice coil= Bl*i     (i=current)
Levc = 0.2           # Voice coil inductance - not provided by manufacturer   
#-----------------------------------------------                            
Qms = 3.06          # Mechanical Q at f0
Qes = 0.7           # Electrical Q at f0
Qts = 0.570         # Total quality factor at f0:   Qts=Qms*Qes/(Qms+Qes) 
Rm = (((2*np.pi)*f0)*Mmd)/Qms
#====================
# Voice coil Impedance (replaces Re) Added 14 Septembre 2007
# Mechanical resistance from Qms definition: (Qms=2*pi*f0*Mmd/Rm) 
#====================
Ze = Re+(Levc*1j)*w
```

```python
#======================================================================
#           SPEAKER  Enclosure Parameters + Electrical Parameters          |
#======================================================================
Vbcc= 500;                        # Enclosure volume (cm3)
Vb=Vbcc*1e-06;
kb = ((ro*(cel**2))*(Sd**2))/Vb;    # Enclosure "stiffness" N.m    
Cab=   Vb/(ro*(cel**2));
#-----------------------------
#Ma = Air Added Mass
#-----------------------------
a = np.sqrt(Sd/np.pi);   # diaphragm radius 
Ma1 = (((8*ro)*Sd)*a)/(3*np.pi);
RAR2=(ro*cel*w**2)/(np.pi*(w**2*a**2 + 2*cel**2));  # dissipation
Rm =   Rm + RAR2*Sd**2
M = Mmd+Ma1;
M=Mmd;
#-----------------------------
# Speaker suspension Stiffness
#-----------------------------
Cas = Vas/(ro*(cel**2));
Cms = Cas/(Sd**2);
ks=1/Cms;
#
ksealed = ks+kb;
#==================================
# Speaker + Enclosure stiffness
#==================================
w0s = np.sqrt(ksealed/M);         # Speaker f0 in the sealed enclosure
f0s = w0s/(2*np.pi);
Qtsenc = np.sqrt(ksealed*M)/(Rm+(Bl**2)/Re);
Cmb=1/kb;
Qtsenc[1]
#============   QTC box and Rs=RAB  Slow Leak in the enclosure ============
Qtc=0.682
Rs= Sd**2./ (2*np.pi*f0s*Qtsenc*Cmb)
#================ RAL Absorption in the ecnlosure ========
Ql=20; 
RAL=Ql/(2*np.pi*f0s*Cmb)
```

```python
#======================================================
#                 PASSIVE RADIATOR CHARACTERISTICS
#======================================================
f0pr=30
Sdpr = 48e-4      # Diaphragm Area (m2)
Vaspr = 3200e-06    # Equivalent volume (m3) - linked to suspension compliance
ksvpr = ((ro*(cel**2))*(Sd**2))/Vaspr # Suspension stiffness  (N/m)
CMpr=1/ksvpr;   # Suspension compliance
Mmdpr= ksvpr/(2*np.pi*f0pr)**2
#-----------------------------------------------
apr = np.sqrt(Sdpr/np.pi)   # passive radiator radius 
Mapr = (((8*ro)*Sdpr)*apr)/(3*np.pi)
#Mmdpr=Mmdpr+ Mapr
#-----------------------------------------------

#-----------------------------------------------                            
Qmspr =10;
RARpr=(ro*cel*w**2)/(np.pi*(w**2*apr**2 + 2*cel**2))  # dissipation through acsoutic radiation
#-----------------------------------------------
Rmpr = (((2*np.pi)*f0pr)*Mmdpr)/Qmspr
Rmpr =   Rmpr + RARpr*Sdpr**2      
```

```python
#=============  Sound Pressure radiated by the speaker in an enclosurezspeaker=  M*(i*w) +  (Rm+ Bl**2/Re)+ ( 1 /(Cms*(i*w)));
Em=2                       # Input Volts  - 2W in 4 Ohms  
pg=Em*Bl/(Sd*Re);          # Newtons/m2
pg=2*pg;                   # "INFINITE BAFFLE": + 6dB  HALF SPACE !  
                           # pg=2*pg   --->  1/(2*np.pi*rr)   half space
                           # pg=pg     --->  1/(4*np.pi*rr)   full space
rr =   1                   # Computation of pressure at radius= 1 meter  Speaker
rrpr=  1                   # Computation of pressure at radius=1 meter  Passive radiator
#==================================================================
zel1=Levc*Sd**2/Bl**2;
zel2=Bl**2/(Re*Sd**2);
#--------------------------

#--------------------------
zspeaker=  M*(1j*w) + Rm + ( 1 /(Cms*(1j*w))) + zel2*Sd**2;
#-----------------------------------
zcavity=( Rs   +  1  /(Cmb*(1j*w)));           #   Rs effect of a leak in the cavity
#----------------------------------------------
# Passive radiator Impedance 
Zspr=Mmdpr*(1j*w) + Rmpr + ( 1 /(CMpr*(1j*w)));  
#----------------------------------------------

zcavity=parallel(zcavity, RAL)  

#------------
Z1=parallel(zcavity,Zspr)          # Enclosure  + Passive Radiator 
Ztotal=zspeaker+Z1                 # Speaker + Enclosure  + Passive Radiator 
Zclosed=series(zspeaker,zcavity)   # Speaker + Sealed Enclosure
#------------
Ud=pg/Ztotal;
Uclosed=pg/Zclosed;
Up1=Ud/(1+Zspr/zcavity);
#------------
#
#============================================================
pud=(ro*Sd**2/(4*np.pi*rr))*np.exp(-(1j *k)*rr  )*1j*w*Ud 
puclose=(ro*Sd**2/(4*np.pi*rr))*np.exp(-(1j *k)*rr  )*1j*w*Uclosed 
pup1=(ro*Sd**2/(4*np.pi*rrpr))*np.exp(-(1j *k)*rrpr  )*1j*w*Up1
#=============================================================
puu=pud-pup1;  

#=============================================================
#  Xd=  ud/jw   displacement = velocity/ jw
#  Ud = ud*Sd  Volume velocity= velocity * Surface area Sd
#  Xd= Ud/(jwSd)
#  conversion Mechanical impedance/ Acoustical impedance Sd**2
#  0.707 factor ?  
#=============================================================
Xclosed=  np.sqrt(2)/2*Uclosed*Sd**2/(1j*w*Sd)  # Displacement speaker diaphragm (closed box)
Xd=       np.sqrt(2)/2*Ud*Sd**2/(1j*w*Sd)       # Displacement speaker diaphragm with PR
Xpr=      np.sqrt(2)/2*Up1*Sd**2/(1j*w*Sdpr)    # Displacement Passive radiator  from pressure radiated
#=============================================================
```

```python
fig = plt.figure(figsize = (10, 6))
plt.semilogx(afreq,20*np.log10(abs(puclose)/0.00002),'b')
plt.semilogx(afreq,20*np.log10(abs(puu)/0.00002),'r')
plt.grid(b=True, which='major', color='grey', linestyle='-')
plt.grid(b=True, which='minor', color='grey', linestyle='--')
plt.ylim(0,100);
plt.xlim(1,1000);
plt.xlabel('Frequency (Hz)');
plt.ylabel('dB SPL');
plt.legend(['Sealed Enclosure', 'Passive radiator']);
plt.title('Frequency Response @ 1 meter. Peerless + Passive Radiator in '+str(Vbcc)+' cc enclosure');
```
